```python
# dataset creation
import pandas as pd

data ={
    "Roll_No.":[101,102,103,104,105],
    "Gender":["Female","Male","Female","Male","Female"],
    "Department":["computer","IT","ENTC","Mechanical","Computer"],
    "CGPA":[8.2,6.4,9.2,9.0,6.7]
}
df = pd.DataFrame(data)
print(df)
# data is transformed in tabular form

# now we will convert this data into csv format
df.to_csv("students.csv",index=False)
print(df.to_csv)
#there won't be indexing in the csv file


# number of rows and columns
print(df.shape)


# number of elements
print(df.size)


# gives full information of the data
df.info()


# does statistical analysis
df.describe()
```
